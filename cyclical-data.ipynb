```python
from matplotlib import pyplot as plt
import numpy as np
hours = [0, 3, 6, 9, 12, 15, 18, 21, 23]
hr_sin = np.sin(np.array(hours) * (np.pi/-12.))
hr_cos = np.cos(np.array(hours) * (np.pi/-12.))

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

plt.plot(hr_cos, hr_sin, '.', data=hours)
plt.xlabel('hr_cos')
plt.ylabel('hr_sin')

for index, xy in enumerate(zip(hr_cos, hr_sin)):
    ax.annotate('%s' % hours[index], xy=xy, textcoords='data')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
hours = [0, 3, 6, 9, 12, 15, 18, 21, 23]
hr_sin = np.sin(np.array(hours) * (np.pi/-12.))
hr_cos = np.cos(np.array(hours) * (np.pi/-12.))

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111)

plt.plot(hr_cos, hr_sin, '.', data=hours, markersize=16)
plt.xlabel('hr_cos')
plt.ylabel('hr_sin')

print_cos = hr_cos * 1.05
print_sin = hr_sin * 1.05
print_cos[2] += 0.04 # hour 6
print_cos[3] -= 0.04 # hour 9

print_sin[4] += 0.04 # hour 12
print_sin[6] -= 0.04 # hour 18
print_cos[6] += 0.04 # hour 18

for index, xy in enumerate(zip(print_cos, print_sin)):
    ax.annotate('%s' % hours[index], xy=xy, textcoords='data')
plt.show()
```
